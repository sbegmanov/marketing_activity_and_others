### Data Models and Structured Data

A data model:
   * Structured Data = completely structured or wellstructured data, flat tabular form, no duplicate columns
   * Semistructured data = variable lengths, different data types (numerical or categorical) in same column,nested or hierarchical tabular structure, columns (attributes)-no duplicates, rows(observations)-may be, may be missing values, stored accurately in NoSQL databases, Apache Parquet files, JSON files, and so on.
   * Unstructured data =  may not be tabular, even if it is tabular, the number of attributes or columns per observation may be completely arbitrary, can be stored as text files, CSV files, Excel files, images, audio clips, same data could be represented in different arrays, the attributes might not match each other, with values leaking into other parts
